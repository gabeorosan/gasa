```python
import yt_dlp
max_results = 10

urls = []

def search_youtube(query, max_results=5):
    ydl_opts = {
        'quiet': True,
        'default_search': 'ytsearch',
        'extract_flat': True,
    }

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        result = ydl.extract_info(f"ytsearch{max_results}:{query}", download=False)
        entries = result.get('entries', [])
        return entries

def check_chinese_subtitles(url):
    ydl_opts = {}

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        info = ydl.extract_info(url, download=False)
        subtitles = info.get('subtitles', {})
        automatic_captions = info.get('automatic_captions', {})

        # Convert dict_keys to lists before concatenating
        all_lang_codes = list(subtitles.keys()) + list(automatic_captions.keys())

        return any('zh' in lang_code for lang_code in all_lang_codes)
queries = [
    "中国古典音乐",  # Chinese Classical Music
    "华语流行歌曲",  # Chinese Pop Songs
    "粤语说唱音乐",  # Cantonese Rap Music
    "中国功夫电影",  # Chinese Kung Fu Movies
    "古装历史剧",    # Historical Costume Dramas
    "普通话学习教程",  # Mandarin Learning Tutorials
    "中国美食烹饪",  # Chinese Cooking and Cuisine
    "北京旅游指南",  # Beijing Travel Guide
    "上海夜生活",    # Shanghai Nightlife
    "中国节庆庆祝",  # Chinese Festivals and Celebrations
    "中国科技创业",  # Chinese Tech Startups
    "深圳电子市场",  # Shenzhen Electronics Market
    "中国足球超级联赛", # Chinese Super League (Football)
    "太极初学教程",  # Tai Chi for Beginners
    "中国当代艺术展",  # Chinese Contemporary Art Exhibitions
    "古代中国诗歌朗诵", # Ancient Chinese Poetry Recitation
    "中国花园和风景",  # Chinese Gardens and Landscapes
    "中国野生动物纪录片", # Wildlife in China Documentaries 
]
urls = []
for q in queries:
    videos = search_youtube(q, max_results=max_results)

    for video in videos:
        urls.append(video['url'])
```

```python
import yt_dlp

URL = 'https://www.youtube.com/watch?v=BaW_jenozKc'

ydl_opts = {}
with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    info = ydl.extract_info(URL, download=False)
    
    title = info['title']
    subtitles = info.get('subtitles', {})
    automatic_captions = info.get('automatic_captions', {})

    print(f"Title: {title}")
    
    chinese_subtitles = any(lang_code.startswith('zh') for lang_code in subtitles.keys())
    chinese_auto_captions = any(lang_code.startswith('zh') for lang_code in automatic_captions.keys())

    if chinese_subtitles or chinese_auto_captions:
        print("Chinese subtitles available!")
    else:
        print("No Chinese subtitles found.")
```

```python
import yt_dlp

def search_youtube(query, max_results=5):
    ydl_opts = {
        'quiet': True,
        'extract_flat': True,
        'max_downloads': max_results,
        'default_search': 'ytsearch',
        'format': 'best',
    }

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        result = ydl.extract_info(query, download=False)
        entries = result.get('entries', [])
        
        if not entries:
            print(f"No entries found for query: {query}")
            print(result)  # Print the entire result to debug

        return entries


categories = ["music", "movies", "education", "technology"]
keywords = ["Chinese", "Mandarin", "subtitles"]



for category in categories:
    for keyword in keywords:
        query = f"{keyword} {category}"
        print(f"Searching for: {query}")
        videos = search_youtube(query)
        
        for video in videos:
            print(f"Title: {video['title']}")
            print(f"URL: https://www.youtube.com/watch?v={video['id']}")
        print()
```
